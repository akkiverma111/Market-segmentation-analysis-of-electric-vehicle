**Importing Dependencies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
```

```python
# Loading the csv data to a Pandas DataFrame

states_data = pd.read_csv('/content/drive/MyDrive/states_data_car.csv')
```

```python
# Get the counts of each subsidy category
subsidy_counts = states_data["subsidy"].value_counts()

# Create a bar plot
sns.countplot(x=states_data["subsidy"])

# Set plot title and labels
plt.title('Count plot of subsidy')
plt.xlabel('Subsidy')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Create the count plot using Seaborn
sns.countplot(x=states_data["road tax"])

# Set plot title and labels
plt.title('Count plot of road tax')
plt.xlabel('Road Tax')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.set_theme(style="whitegrid")
sns.barplot(x='state', y='petrol', data=states_data, color='lightslategrey')
plt.title('Petrol Price in Different States', fontsize=16)
plt.ylabel('Petrol Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.set_theme(style="whitegrid")
sns.barplot(x='state', y='diesel', data=states_data, color='lightslategrey')
plt.title('Diesel Price at different states', fontsize=10)
plt.ylabel('Diesel Price')
plt.xlabel('State Name')
plt.xticks(rotation=90)
plt.show()
```
